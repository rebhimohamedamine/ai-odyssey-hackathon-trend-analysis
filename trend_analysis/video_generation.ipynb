```python
model_id = "stabilityai/stable-diffusion-2-1"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
topic = "anime, edit , anime"  # Replace with your trending topic
duration = 10  # Duration of the video in seconds
fps = 12  # Frames per second
```

```python
# Calculate the number of frames
num_frames = duration * fps
```

```python
os.makedirs("frames", exist_ok=True)
```

```python
prompt_template = (
    "A TikTok video about {topic}, cinematic lighting, vibrant colors, 4K resolution, "
    "trending on social media, engaging and eye-catching, perfect for TikTok"
)
prompt = prompt_template.format(topic=topic)
```

```python
for i in range(num_frames):
    # Adjust the prompt slightly for each frame (optional)
    frame_prompt = f"{prompt}, frame {i+1}/{num_frames}"
    
    # Generate the image
    image = pipe(frame_prompt).images[0]
    
    # Save the frame explicitly
    frame_path = f"frames/frame_{i:04d}.png"
    image.save(frame_path)
    print(f"Saved {frame_path}")
```
