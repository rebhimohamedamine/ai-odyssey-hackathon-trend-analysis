```python
import requests

# TikTok API Parameters
ACCESS_TOKEN = "YOUR_ACCESS_TOKEN"
UPLOAD_ENDPOINT = "https://open-api.tiktok.com/post/video/upload/"

# Upload the video to TikTok
def post_on_tiktok(video_path, caption):
    with open(video_path, "rb") as video_file:
        files = {"video": video_file}
        payload = {"access_token": ACCESS_TOKEN, "caption": caption}
        response = requests.post(UPLOAD_ENDPOINT, files=files, data=payload)

    if response.status_code == 200:
        print("Successfully posted on TikTok!")
    else:
        print(f"TikTok Error: {response.text}")

# Example usage
post_on_tiktok(video_path, "New trend with #AI #Viral ")
```
